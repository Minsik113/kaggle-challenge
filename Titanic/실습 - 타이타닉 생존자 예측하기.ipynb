# 3. 학습하기 위해 데이터를 정리하고, 특징을 선택하기

```python
raw_data.head(3)
```

```python
tmp = []
# 
for sex in raw_data['sex']:
    if sex == 'female':
        tmp.append(0)
    elif sex == 'male':
        tmp.append(1)
    else:
        tmp.append(np.nan)
raw_data['sex'] = tmp
raw_data.head(3)
```

```python
# 정보가 없는 NAN을 빼고 다시 저장 - 데이터가 263개 줄어듦
raw_data.shape # 1309 x 15
raw_data = raw_data[raw_data['age'].notnull()]
raw_data = raw_data[raw_data['sibsp'].notnull()]
raw_data = raw_data[raw_data['parch'].notnull()]
raw_data = raw_data[raw_data['fare'].notnull()]
raw_data.shape # 1045 x 15
raw_data.info()
```

# 4. 생존자 예측을 위한 모델 수립

```python
train_pre = raw_data[['pclass','sex','age','sibsp','parch','fare']]
train_pre.head(3)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_pre
                                                   ,raw_data[['survived']]
                                                   ,test_size=0.1
                                                   ,random_state=13)
X_train.shape # 940 x 6
X_train.head(3) 
```

## 4-1. 인덱스가 인전테이블의 num가지고옴 -> 초기화해주자

```python
X_train = X_train.reset_index()
X_train = X_train.drop(['index'], axis=1)
X_train.head(3)
```

```python
X_test = X_test.reset_index()
X_test = X_test.drop(['index'], axis=1)
X_test.head(3)
```

```python
y_train = y_train.reset_index()
y_train = y_train.drop(['index'], axis=1)
y_train.head(3)
```

```python
y_test = y_test.reset_index()
y_test = y_test.drop(['index'], axis=1)
y_test.head(3)
```

## 4-2. 모델은 결정트리 이용

```python
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier(max_depth=3, random_state=13)
tree_clf.fit(X_train, y_train)

print(f'decisiontree모델 학습 score: {tree_clf.score(X_train,y_train) : .2f}') # 0.81
```

```python
X_train.columns
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree_clf
               ,out_file='titanic.dot' # tree.dot이라고 파일을 만듦
               ,class_names=['died','survived']
               ,feature_names=X_train.columns
               ,impurity=False
               ,filled=True) 

import graphviz
with open(("titanic.dot"), "r", encoding='utf-8') as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph))
```

```python
from sklearn.metrics import accuracy_score
y_pred = tree_clf.predict(X_test)
print(f"Test 정확도: {accuracy_score(y_test,y_pred) * 100}") # 84.76
```

# 5. 디카프리오 생존 예측해보기

```python
# pclass, sex, age, sibsp, parch, fare
dicaprio = [3,1,19,0,0,5] # 3등석, 남, 나이19, 형제자매없음, 부모님없이혼자탐, 5달러에탐
winslet = [1,0,17,1,1,100]
```

```python
print(f'dicaprio가 살확률 {tree_clf.predict_proba([dicaprio])}') # 0.125
print(f'winslet이 살확률 {tree_clf.predict_proba([winslet])}') # 0.975
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, y_train)
print(f'knn모델 학습 score: {knn_clf.score(X_train, y_train)}') #0.77
y_pred = knn_clf.predict(X_test)
print(f"Test 정확도: {accuracy_score(y_test,y_pred) * 100}") # 73.3

Ran_clf = RandomForestClassifier(n_estimators=13)
Ran_clf.fit(X_train, y_train)
print(f'randomforest모델 학습 score: {Ran_clf.score(X_train, y_train)}') #0.97
y_pred = Ran_clf.predict(X_test)
print(f"Test 정확도: {accuracy_score(y_test,y_pred) * 100}") # 78.09
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
kf_tree_clf = DecisionTreeClassifier()
score = cross_val_score(kf_tree_clf
                       ,X_train
                       ,y_train
                       ,cv=k_fold
                       ,scoring='accuracy')
print(score)
print(f'k_fold + decisiontree모델 학습 score {round(np.mean(score)*100, 2)}') # 0.74
```
